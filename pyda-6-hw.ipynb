## Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

movies = pd.read_csv('./movies/movies.csv')
ratings = pd.read_csv('./movies/ratings.csv')
```

```python
# just showing of data structure
# common column is movieId in two datasets
movies.head()
```

```python
ratings.head()
```

```python
(
    ratings
    .merge(movies, on='movieId')
    .query('rating == 5.0')
    [['title', 'rating']]
    .groupby(by='title')
    .size()
    .sort_values(ascending=False)
    .head(1)
)
```

## Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import pandas as pd

p_df = pd.read_csv('./power.csv')

(
    p_df
    .query('country == "Latvia" or country == "Lithuania" or country == "Estonia"')
    .query('category == 4 or category == 12 or category == 21')
    .query('2005 <= year <= 2010')
    .query('quantity > 0')
    ['quantity']
    .sum()
)
```
